```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("/Sample - Superstore.csv", encoding='latin-1')

# Display first rows
print("Dataset Preview:")
print(df.head())

# Basic info
print("\nDataset Info:")
print(df.info())

# Check missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Total sales and profit
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()

print("\nTotal Sales:", total_sales)
print("Total Profit:", total_profit)

# Category-wise sales and profit
category_analysis = df.groupby('Category')[['Sales', 'Profit']].sum()
print("\nCategory-wise Analysis:")
print(category_analysis)

# Region-wise profit
region_profit = df.groupby('Region')['Profit'].sum()
print("\nRegion-wise Profit:")
print(region_profit)

# Monthly sales trend
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Month'] = df['Order Date'].dt.month

monthly_sales = df.groupby('Month')['Sales'].sum()

# Plot: Monthly Sales Trend
plt.figure()
monthly_sales.plot()
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()

# Plot: Category-wise Sales
plt.figure()
sns.barplot(x=category_analysis.index, y=category_analysis['Sales'])
plt.title("Sales by Category")
plt.show()

# Plot: Region-wise Profit
plt.figure()
sns.barplot(x=region_profit.index, y=region_profit.values)
plt.title("Profit by Region")
plt.show()
```
